# NBA Wins Prediction (PM1)

This notebook trains regression models to predict NBA team wins from advanced team statistics.


### TPOTs fix and Imports

```python
# Force TPOTs to stop the whole DASK problem
import os, sys
os.environ["TPOT_NO_DASK"] = "1"

print("Python:", sys.version)
print("Executable:", sys.executable)

import warnings
warnings.filterwarnings(
    "ignore",
    message="X does not have valid feature names*"
)
```

```python
import pandas as pd
import numpy as np
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import HistGradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from tpot import TPOTRegressor
import joblib
import matplotlib.pyplot as plt
```

## Load data and define y


```python
# Load data
df = pd.read_csv("advanced_stats_total.csv")

y = pd.to_numeric(df["W"], errors="coerce")
years = pd.to_numeric(df["Year"], errors="coerce")
```

### Clean data

```python
drop_cols = [
    "Unnamed: 0", "Rk", "W", "L", "PW", "PL",
    "SRS", "MOV", "SOS",
    "Arena", "Attend", "Attend/G",
    "...29", "...30", "...31",
    "Team",
    "eFG_percent_offense",
    "TOV_percent_defense",
]

X = df.drop(columns=drop_cols, errors="ignore").copy()
X = X.apply(pd.to_numeric, errors="coerce")
years = pd.to_numeric(df["Year"], errors="coerce")


print("X dtypes:\n", X.dtypes)
print("Any object columns left?", (X.dtypes == "object").any())
print("NaNs in X (max per col):", X.isna().sum().max())
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Remove unusable rows
mask = y.notna() & years.notna()
X = X.loc[mask].reset_index(drop=True)
y = y.loc[mask].reset_index(drop=True)
years = years.loc[mask].reset_index(drop=True)

y.describe()
```

## Train/test split


```python
# Split train/test data by year
train_idx = years <= 2019
test_idx = years >= 2020

X_train, X_test = X.loc[train_idx].copy(), X.loc[test_idx].copy()
y_train, y_test = y.loc[train_idx].copy(), y.loc[test_idx].copy()
years_train, years_test = years.loc[train_idx].copy(), years.loc[test_idx].copy()

print("Train rows:", len(X_train), "| Test rows:", len(X_test))
print("Train years:", int(years_train.min()), "-", int(years_train.max()))
print("Test years :", int(years_test.min()), "-", int(years_test.max()))
```

### Fill missing values

```python
train_meds = X_train.median(numeric_only=True)
X_train = X_train.fillna(train_meds)
X_test = X_test.fillna(train_meds)

print("NaNs left train:", X_train.isna().sum().sum())
print("NaNs left test :", X_test.isna().sum().sum())
```

## Evaluation baselines


```python
# Make an quick eval for the regression models
def eval_reg(model, X_train, y_train, X_test, y_test, threshold=3, label="±3 wins"):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    mae  = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    r2   = r2_score(y_test, pred)

    within = np.mean(np.abs(y_test - pred) <= threshold)

    print(model.__class__.__name__)
    print(f"MAE:  {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R^2:  {r2:.3f}")
    print(f"% within {label}: {within*100:.1f}%\n")
```

```python
# Baseline Regressor
dummy = DummyRegressor(strategy="mean")
eval_reg(dummy, X_train, y_train, X_test, y_test, threshold=3, label="±3 wins")
```

```python
# Standard Linear Regressor
ridge = Ridge(alpha=1.0, random_state=42)
eval_reg(ridge, X_train, y_train, X_test, y_test, threshold=3, label="±3 wins")
```

## XGBoost


```python


xgb = XGBRegressor(
    n_estimators=800,
    learning_rate=0.03,
    max_depth=4,
    subsample=0.9,
    colsample_bytree=0.9,
    reg_lambda=1.0,
    random_state=42,
    n_jobs=-1
)

eval_reg(xgb, X_train, y_train, X_test, y_test, threshold=3, label="±3 wins")
```

## RandomizedSearchCV


```python
# Order training data by year
order = np.argsort(years_train.values)
X_train_ord = X_train.iloc[order].reset_index(drop=True)
y_train_ord = y_train.iloc[order].reset_index(drop=True)

tscv = TimeSeriesSplit(n_splits=5)

# Params to search
param_dist = {
    "n_estimators": [300, 500, 800, 1200, 1600],
    "learning_rate": np.linspace(0.01, 0.12, 12),
    "max_depth": [2, 3, 4, 5, 6],
    "min_child_weight": [1, 2, 3, 5, 8],
    "subsample": np.linspace(0.6, 1.0, 5),
    "colsample_bytree": np.linspace(0.6, 1.0, 5),
    "gamma": [0, 0.05, 0.1, 0.2, 0.4],
    "reg_alpha": [0, 1e-4, 1e-3, 1e-2, 0.1, 1.0],
    "reg_lambda": [0.5, 1.0, 1.5, 2.0, 3.0],
}

base_xgb = XGBRegressor(
    objective="reg:squarederror",
    random_state=42,
    n_jobs=-1
)

search = RandomizedSearchCV(
    estimator=base_xgb,
    param_distributions=param_dist,
    n_iter=400,
    scoring="neg_mean_absolute_error",
    cv=tscv,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Fit to ordered training data
search.fit(X_train_ord, y_train_ord)

print("Best CV MAE:", -search.best_score_)
print("Best params:")
for k, v in search.best_params_.items():
    print(f"  {k}: {v}")
```

```python
best_xgb = search.best_estimator_

eval_reg(best_xgb, X_train, y_train, X_test, y_test)

pred = best_xgb.predict(X_test)
```

```python
joblib.dump(best_xgb, "nba_wins_xgb_model.joblib")
print("Saved:", "nba_wins_xgb_model.joblib")
```

## TPOT


```python


tpot = TPOTRegressor(
    generations=5,
    population_size=20,
    cv=5,
    random_state=42,
    n_jobs=8
)

tpot.fit(X_train, y_train)
```

```python
best_pipe = tpot.fitted_pipeline_
best_pipe
```

```python
eval_reg(best_pipe, X_train, y_train, X_test, y_test, threshold=3, label="±3 wins")
```

```python

joblib.dump(best_pipe, "nba_wins_tpot_model.joblib")
print("Saved:", "nba_wins_tpot_model.joblib")
```

```python
best_model = best_pipe
best_label = "TPOT"

pred = best_pipe.predict(X_test)
```

### Summary metrics and results

```python

abs_err = np.abs(pred - y_test.values)

within_1 = (abs_err <= 1).mean()
within_2 = (abs_err <= 2).mean()
within_3 = (abs_err <= 3).mean()
within_5 = (abs_err <= 5).mean()

print(f"% within ±1 win: {within_1*100:.1f}%")
print(f"% within ±2 wins: {within_2*100:.1f}%")
print(f"% within ±3 wins: {within_3*100:.1f}%")
print(f"% within ±5 wins: {within_5*100:.1f}%")
```

```python
meta = df.loc[mask, ["Team", "Year"]].reset_index(drop=True)

results = pd.DataFrame({
    "Team": meta["Team"].astype(str),
    "Year": pd.to_numeric(meta["Year"], errors="coerce"),
    "actual_wins": y_test.values,
    "predicted_wins": pred
}, index=X_test.index)

results["error"] = results["predicted_wins"] - results["actual_wins"]
results["abs_error"] = results["error"].abs()

# Clean team names (if they include trailing * etc.)
results["Team"] = (
    results["Team"]
    .str.replace(r"\*", "", regex=True)
    .str.strip()
)
```

## Predicted vs Actual Graph

```python

worst = results.sort_values('abs_error', ascending=False).head(5)

plt.figure(figsize=(8, 8))
plt.scatter(results['actual_wins'], results['predicted_wins'], alpha=0.6)

plt.plot(
    [results['actual_wins'].min(), results['actual_wins'].max()],
    [results['actual_wins'].min(), results['actual_wins'].max()],
    linestyle='--'
)

for _, row in worst.iterrows():
    plt.annotate(
        f"{row['Team']} {row['Year']}",
        xy=(row["actual_wins"], row["predicted_wins"]),      
        xytext=(-80, 10),                                     
        textcoords="offset points",
        fontsize=8,
        arrowprops=dict(
            arrowstyle="->",
            linewidth=0.8,
            alpha=0.8
        )
    )

plt.xlabel("Actual Wins")
plt.ylabel("Predicted Wins")
plt.title("Largest Prediction Errors (Annotated)")

plt.show()
```

### Final Team x Year table

```python
subset = results.loc[
    results["Year"].between(2020, 2023),
    ["Team", "Year", "actual_wins", "predicted_wins"]
].copy()

subset = subset.sort_values(["Team", "Year"])
subset["actual_wins"] = subset["actual_wins"].round(1)
subset["predicted_wins"] = subset["predicted_wins"].round(1)

for team, g in subset.groupby("Team"):
    display(
        g.set_index(["Team", "Year"]).sort_index()
    )
```
